```python
#draw a line plot
kashti = sns.load_dataset("titanic")

#sns.set_style('darkgrid')
sns.barplot(x = "sex", y = "alone", data = kashti, hue= "who", order=("female", "male")) #oder means which one you want first

#sns.lineplot(x = "species", y = "sepal_width", data = phool)
#change figure
plt.title("Titanic ka plot")
plt.show()
```

```python
#add char color
#draw a line plot
kashti = sns.load_dataset("titanic")

#sns.set_style('darkgrid')
sns.barplot(x = "sex", y = "alone", data = kashti, hue= "who", order=("female", "male"), color = "blue") #oder means which one you want first

#sns.lineplot(x = "species", y = "sepal_width", data = phool)
#change figure
plt.title("Titanic ka plot")
plt.show()
```

```python
#remove bar inter line
#add char color
#draw a line plot
kashti = sns.load_dataset("titanic")

#sns.set_style('darkgrid')
sns.barplot(x = "sex", y = "alone", data = kashti, hue= "who", order=("female", "male"), color = "blue", ci = None) #oder means which one you want first

#sns.lineplot(x = "species", y = "sepal_width", data = phool)
#change figure
plt.title("Titanic ka plot")
plt.show()
```

```python
#add color pelete i.e combiation of color which dim dim and dim
#remove bar inter line
#add char color
#draw a line plot
kashti = sns.load_dataset("titanic")

#sns.set_style('darkgrid')
sns.barplot(x = "sex", y = "alone", data = kashti, hue= "who", order=("female", "male"), color = "blue", ci = None,
           palette="pastel") #oder means which one you want first

#sns.lineplot(x = "species", y = "sepal_width", data = phool)
#change figure
plt.title("Titanic ka plot")
plt.show()
```
