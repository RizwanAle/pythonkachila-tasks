### **Data Normalization**
- Uniform the data
- They have same impact
- Two fish are very far from each other(not suiable in data)

#### Normalize data 'parametic test' 
#### Not normalize 'non parametric test'
##### normal data which makes bell curve


```python
ks = ks1[['age in days', 'fare']]
ks
```

- The above data is really in wide range we need to normalize because hard to compare 
- Normalization change the values to the range of 0 to 1(to emit wide range)

#### **Method of normalization**
    - Simle feature scaling 
      - x(new)=x(old) / x(max)
    - Min-Max method
    - Z-score (standard score) -3 to +3
    - Log Tranformation 

```python
#simple feature scaling 
ks['fare'] = ks1['fare']/ks1['fare'].max()
ks
```

```python
# now change ages wide range into short range
ks['age in days'] = ks1['age in days']/ks1['age in days'].max()
ks
```

```python
# Min - Max method
ks['fare'] = (ks['fare']-ks['fare'].min()) / ((ks['fare']-ks['fare'].max()-ks['fare'].min())) 
ks.head()
```

### Z-Score Method

```python
ks['fare'] = (ks['fare']-ks['fare'].mean()) / ks['fare'].std()
ks.head
```
