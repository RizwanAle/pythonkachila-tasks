```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
df = pd.read_csv("ml_data_salary.csv")
df.head()
```

```python
X = df[["age", "distance", "YearsExperience"]]
y = df["Salary"]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = X.values
model = model.fit(X, y)
model
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[31.1,77.74,1.1]])
```
